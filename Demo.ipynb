```python
import pandas as pd

data = {
    "name": ["sanju", "riya","ahil", "nitin", "josu"],
    "math": [98,68,54,78,99],
    "science":[56,58,12,45,75],
    "english":[98,68,34,58,22]
}

df= pd.DataFrame(data)
df.to_csv("student_marks.csv", index = False)
print("csv file created sucessfully")
df["math"].fillna(df["math"].mean(), inplace=True)
df["science"].fillna(df["science"].mean(), inplace=True)
df
```

```python
df["grade"] = ["A", "B","A","C","D"]
df.groupby("grade")[["math","science","english"]].mean()
```

```python
df.to_csv("new_studentfile.csv", index=False)
print("done sucessfully")
```

```python
top_students = df.nlargest(2, "total")
print("top students", top_students)

bottom_students =df.nsmallest(2, "total")
print("bottom students", bottom_students)
```

```python
sub_avg = df[["math","science","english"]].mean()
print("\nsubject is avrage\n", sub_avg)
```

```python
sub_avg.plot(kind="bar", title="average subject marks", color=["skyblue", "lightgreen", "salmon"])
```
